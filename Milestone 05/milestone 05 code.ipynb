## Section D: Task 2 – Calculating Basic Probability

```python
import pandas as pd
# Assume df is loaded and column names are cleaned
df = pd.read_csv('academic stress level.csv')
df.columns = df.columns.str.strip()

# Total number of outcomes (N)
N = len(df)

# Define Events (A, B, C) again with cleaned column names
A = df[df['Your Academic Stage'] == 'undergraduate']
B = df[df['Rate your academic stress index'] == 5]
C = df[df['Peer pressure'] > 3]

# 1. Compute empirical probability
P_A = len(A) / N
P_B = len(B) / N
P_C = len(C) / N

print(f"Total observations (N): {N}\n")
print(f"P(A) = {len(A)} / {N} = {P_A:.4f}")
print(f"P(B) = {len(B)} / {N} = {P_B:.4f}")
print(f"P(C) = {len(C)} / {N} = {P_C:.4f}")

# 2. Verification
print("\n--- Probability Verification ---")
print(f"Is P(A) between 0 and 1? {0 <= P_A <= 1}")
print(f"Is P(B) between 0 and 1? {0 <= P_B <= 1}")
print(f"Is P(C) between 0 and 1? {0 <= P_C <= 1}")
```

## Section E: Task 3 – Combined Events

```python
import pandas as pd
import numpy as np
df = pd.read_csv('academic stress level.csv')
df.columns = df.columns.str.strip()
N = len(df)
A = df[df['Your Academic Stage'] == 'undergraduate']
B = df[df['Rate your academic stress index'] == 5]
P_A = len(A) / N
P_B = len(B) / N
print(f"Total observations (N): {N}\n")
A_int_B = df[(df['Your Academic Stage'] == 'undergraduate') & (df['Rate your academic stress index'] == 5)]
P_A_int_B = len(A_int_B) / N
print('A ∩ B size:', len(A_int_B))
print(f'P(A ∩ B) = {P_A_int_B:.4f}')
A_union_B = df[(df['Your Academic Stage'] == 'undergraduate') | (df['Rate your academic stress index'] == 5)]
P_A_union_B = len(A_union_B) / N
print('\nA ∪ B size:', len(A_union_B))
print(f'P(A ∪ B) = {P_A_union_B:.4f}')
A_comp = df[df['Your Academic Stage'] != 'undergraduate']
P_A_comp = len(A_comp) / N
print('\nAᶜ size:', len(A_comp))
print(f'P(Aᶜ) = {P_A_comp:.4f}')
rule_value = P_A + P_B - P_A_int_B

print('\n--- Verification of Addition Rule ---')
print(f'P(A) + P(B) - P(A ∩ B) = {rule_value:.4f}')
print(f'Actual P(A ∪ B): {P_A_union_B:.4f}')
print(f'Verification result: {np.isclose(rule_value, P_A_union_B)}')
```

## Section F: Visualization

```python

counts = [len(A), len(A_comp)]
labels = [f'A: Undergraduate ({len(A)})', f'Aᶜ: Not Undergraduate ({len(A_comp)})']

plt.figure(figsize=(7, 5))
plt.bar(labels, counts, color=['skyblue', 'lightcoral'])
plt.title('Frequency of Academic Stage (Event A vs. Aᶜ)')
plt.ylabel('Count')
plt.xlabel('Event Outcome')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('task4_plot1_event_A.png')
```

##  Section G: Reflection and Conclusion

It's been great exploring the concept of probability using this dataset! By treating our $N=140$ observations as the full sample space, we've moved beyond simple counts and started quantifying the likelihood of different student experiences. The results offer some clear, data-driven insights into the academic environment.



The standout finding here is that our sample is overwhelmingly focused on one group. Event A (Student is an undergraduate) was, by far, the most probable event with $P(A) \approx 0.7143$. This just confirms that we should keep the undergraduate experience top-of-mind when interpreting any other results.

Looking specifically at the pressure ratings, Event C (Peer Pressure greater than 3) was more likely, occurring with $P(C) \approx 0.3286$. This is significantly higher than the probability of an individual student hitting the absolute peak stress index (Event B).

We successfully verified the Addition Rule of Probability, showing that the likelihood of a student being either an undergraduate or having maximum stress ($P(A \cup B)$) is indeed the sum of their individual probabilities minus their overlap, proving that our empirical calculations align perfectly with theory!


The most striking pattern isn't in an individual event, but in the cumulative picture of stress. While $P(\text{Stress Index}=5)$ was notable, the true insight came from combining the highest stress categories.

We found that the probability of a student reporting a high stress index ($\ge 4$) is $P(\ge 4) \approx 0.536$.

This is huge! It means that more than half of the students surveyed are operating under severe academic stress. This strong skew towards the high end of the stress scale is the most significant takeaway and points to a critical well-being issue within the student population.
